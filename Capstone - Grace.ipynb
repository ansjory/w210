## Build out Recommender - Find Similar Questions

```python
# questions

raw_documents= questions['title'] + questions['tags']
```

```python
raw_documents = raw_documents[0:10000]
```

```python
questions = questions[0:10000]
```

```python
print("Number of Questions:",len(raw_documents))

#Tokenizing data
gen_docs = [[w.lower() for w in word_tokenize(text)] 
            for text in raw_documents]

# Create dictionary
dictionary = gensim.corpora.Dictionary(gen_docs)

# Creat Document-Term Matrix
corpus = [dictionary.doc2bow(gen_doc) for gen_doc in gen_docs]

# Creat TF-IDF Model
tf_idf = gensim.models.TfidfModel(corpus)

# Creat Similarity Checker
similar_qs = gensim.similarities.Similarity("",tf_idf[corpus],num_features=len(dictionary))

print("Processing Completed!")
```

```python
Query = 'How to build a bar chart'
Query
```

```python
Query = 'How to build a pie chart based on R and Tableau'
Query
```

```python
query_doc = [w.lower() for w in word_tokenize(Query)]
query_doc_bow = dictionary.doc2bow(query_doc)
query_doc_tf_idf = tf_idf[query_doc_bow]
```

```python
q_sim=similar_qs[query_doc_tf_idf]
```

```python
sim_threshold=0.10
```

```python
# Display similar questions from the past:

questions['Similarity']=q_sim
ques=questions[questions['Similarity']>=sim_threshold]
ques=questions.sort_values('Similarity',ascending=False)
ques.head()
```
